```python
def load_fluencesweep(folderpath, regex):
    
    fns = os.listdir(folderpath)
    
    
    fluences = []
    traces = []

    for fn in fns:
        m = re.search(regex, fn)
        fp  = os.path.join(folderpath,fn)
        if m is not None:
            p = os.path.join(folderpath,fn)
            fluences.append(float(m.groups()[1]))
            traces.append(load.load_trace(fp))

    df = pd.DataFrame(index = traces[0].index, columns = fluences)

    for i, trace, in enumerate(traces):
        df[fluences[i]] = trace

    return df
```
