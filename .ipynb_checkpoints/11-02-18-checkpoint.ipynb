```python
#Correct for quartz
T_corr = T.copy()
for samp in T_corr:
    T_corr[samp] = T[samp]/quartz
    
#normalize to maximum value    
T_norm = T_corr.copy()
for samp in T_norm:
    T_norm[samp] = T_corr[samp]/T_norm[samp].max()

fig, axes = plt.subplots(2,1, figsize = (7,8), sharex = True)

for samp in T.columns:
#     axes[0].plot(T[samp], label = samp)
    axes[0].plot(T_corr[samp], label = labels[samp])
    axes[1].plot(T_norm[samp], label = labels[samp])
    print("Transmittance ", samp, ': ', str(T_corr[samp][532.0]))

# axes[0].plot(quartz, label = 'Quartz')
# axes[0].set_title('Raw Data')
axes[0].set_title('Corrected for Quartz')
axes[1].set_title('Normalized')

axes[0].set_ylabel('Transmission')
axes[1].legend()
axes[1].set_xlabel('Wavelength (nm)')

for ax in axes:
    
    ax.set_yscale('log')
```

```python
#Absorbance
A = T_corr.copy()
for samp in T_corr:
    A[samp] = 1-T_corr[samp]
    
#subtraced absorbance
A_sub = A.copy()
for samp in A:
    A_sub[samp] = A[samp]-(A[samp].max()-1)


fig, axes = plt.subplots(1,2, figsize = (10,3))

FAs = {}

for samp in T.columns:
    axes[0].plot(A[samp], label = samp)
    axes[1].plot(A_sub[samp], label = samp)
    FAs[samp] = A[samp][532.0]
    print("Absorbance ", samp, ': ', str(FAs[samp]))

# axes[0].plot(quartz, label = 'Quartz')
axes[0].set_title('Absorbance')
axes[1].set_title('Subtracting largest absorbance')
# axes[2].set_title('Corrected and Normalized to max')
    
axes[0].set_ylabel('Absorption')
for ax in axes:
    ax.set_xlabel('Wavelength (nm)')
    ax.legend()
#     ax.set_yscale('log')
```

# half-life calculation

```python
sers_hl = {}
# sample = 'A'

# from matplotlib.ticker.ScalarFormatter import ScalarFormatter

for i, sample in enumerate(dfs_cond):
    df = trmc.offsettime(dfs_cond[sample])
    df[df.columns[::-1]]
    df = df.iloc[0:1000,:]
    times = df.index
    df_max = df.max()
    t_hls = []
    fluences = df_max.index
    for fluence in fluences:
        maximum = df_max[fluence]
        time_max = abs(df[fluence] - maximum).idxmin()
        time_half = abs(df[fluence] - maximum/2).idxmin()
        t_hls.append(time_half-time_max)
    sers_hl[sample] = pd.Series(t_hls, index = df_max.index)
    
    
```

```python
fig, axes = plt.subplots(1,4, figsize = (15,5), sharey = True)

dofit_r1 = {'A':True,'B':True,'C': True,'D': True } 
dofit_r2 = {'A':True,'B':True,'C': False,'D': False } 

tau1_hp = pd.Series(index = dofit_r1.keys())
tau2_hp = pd.Series(index = dofit_r1.keys())

sampindex = {'A':0,'B':0,'C': 0,'D': 0 } 

for i, sample in enumerate(dfs_cond):
    df = dfs_cond[sample]
    df = df/df.max()
    c_exp = df[df.columns[sampindex[sample]]]
    
    ax = axes[i]
    ax.plot(c_exp, marker = 'o', label = "")
    
    t_r1_1 = 10e-9
    t_r1_2 = 50e-9
    if(dofit_r1[sample]):
        p = np.polyfit(c_exp[t_r1_1:t_r1_2].index,np.log(abs(c_exp[t_r1_1:t_r1_2])),1)
        fitline =  np.exp(p[1]+ p[0]*c_exp.index)
        tau = abs(1/p[0])
        ax.plot(c_exp.index,fitline,label = ("$\\tau _1$ = " + str("{:0.2E}".format(tau).replace("E-0", "E-") + " s" )))
        tau1_hp[sample] = tau
    else:
        tau1_hp[sample] = np.nan

    
    t_r2_1 = 50e-9
    t_r2_2 = 200e-9
    if(dofit_r2[sample]):
        p = np.polyfit(c_exp[t_r2_1:t_r2_2].index,np.log(abs(c_exp[t_r2_1:t_r2_2])),1)
        fitline =  np.exp(p[1]+ p[0]*c_exp.index)
        tau = abs(1/p[0])
        ax.plot(c_exp.index,fitline,label = ("$\\tau _2$ = " + str("{:0.2E}".format(tau).replace("E-0", "E-") + " s" )))
        tau2_hp[sample] = tau
    else:
        tau2_hp[sample] = np.nan
    
    plt.xticks(rotation=45)
    ax.set_xlim(-50e-9,.45e-6)
    ax.set_xlabel('Time (ns)')
    expf = exp_formatter(-9)
    ax.xaxis.set_major_formatter(FuncFormatter(expf.format_func))
    ax.xaxis.get_offset_text().set_visible(False)
    ax.set_title(labels[sample] + "\nFluence: " + "{:0.2E}".format(df.columns[sampindex[sample]]))
    ax.legend(loc="upper right")
    
axes[0].set_yscale('log')
axes[0].set_ylabel('Normalized $\Delta G$')
axes[0].set_ylim(2e-3,5) 

fig.suptitle('High Power exponetial fitting')
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
n = {}
for sample in dfs_cond:
    I0 = dfs_cond[sample].columns[0]
    power, dng = kin.calc_pow(t,I0,params_model)
    n[sample] = kin.calc_n(dng,**ks[sample])
```
