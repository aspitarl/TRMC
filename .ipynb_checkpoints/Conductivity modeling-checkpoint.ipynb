Drude Model

$\sigma(\omega)$ = $\frac{1}{(1+\omega^2\tau^2)}(1 - i\omega\tau)$

```python
import numpy as np

tau = 1e-5
```

```python
import pandas as pd
```

```python
w = np.logspace(1,10,100)


pre = 1/(1+(tau**2)*(w**2))

pre = pd.Series(pre, index = w)
pre.index.name = 'Frequency'
pre.name = 'conductivity'

sigr = pre

sigi = -1*pre*w*tau
```

```python
import matplotlib.pyplot as plt

sigr.plot(label = 'real')
abs(sigi).plot(label = 'abs(im)')
plt.xscale('log')
plt.yscale('log')
plt.ylabel('conductivity')
plt.legend()
```

```python
ratio = -1*sigr/sigi
ratio.plot()
plt.xscale('log')
plt.yscale('log')
plt.ylabel('ratio')
```

```python
mag = sigr**2 +sigi**2
```

```python
mag.plot()
plt.xscale('log')
plt.ylabel('magnitude')
```
