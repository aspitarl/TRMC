```python
import pandas as pd
import numpy as np
import easygui
import scipy.optimize
import re
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter, FuncFormatter

import kin
import trmc
import os
mpl.rc('font',**{'size' : 12})

class exp_formatter(): 
    """used to format exponentials of ticks"""
    def __init__(self,exponent):
        self.exponent = exponent
    
    def format_func(self,value, tick_number):
        return ("{:0=1.0f}").format(value/10**self.exponent)
#         return ("{:0=1.0f}e" + str(self.exponent)).format(value/10**self.exponent)

defaultpath = 'C:\\Users\\aspit\\OneDrive\\Data\\TRMC\\Sample Set 3\\FrequencySweep2\\20181205\\'

samples = {'kcs': 'Sample KCs' ,'a': 'Sample A_R2' }
```

```python
def fit_lor(sweep, p0, window = 105):
    
    xdata = sweep.index.values
    ydata = sweep.values
    
    minidx = ydata.argmin()
    minfreq = xdata[minidx]
    
    bounds = ([0,0,0],[np.inf,np.inf,np.inf])
    
    sl = slice(minidx-window,minidx+window)
    popt,popc = scipy.optimize.curve_fit(trmc.lor,xdata[sl],ydata[sl], p0 , bounds = bounds)
    fit_func = trmc.lor(xdata,*popt)
    return fit_func, popt, sl
```

```python
fps_sw = {}

fps_sw['KCs_0ms'] = defaultpath + "Sample KCs\\Sweep_0ms_exp.csv"
fps_sw['KCs_1000ms'] = defaultpath + "Sample KCs\\Sweep_1000ms_exp.csv"

fps_sw['A_0ms'] = defaultpath + "Sample A_R2\\Sweep_0ms_exp.csv"
fps_sw['A_1000ms'] = defaultpath + "Sample A_R2\\Sweep_1000ms_exp.csv"

swps = []
maxV = {}
for samp in fps_sw:
    if 'KCs' in samp:
        df = pd.read_csv(fps_sw[samp], index_col = 0)
        swps.append(df['Experimental R'])
        maxV[samp] = df[' Vsignal(V)'].max()
    else:
        df = pd.read_csv(fps_sw[samp], index_col = 0, skiprows = 4)
        swps.append(df['Reflectivity'])
        maxV[samp] = 0.12

    
    
swps = dict(zip(fps_sw.keys(),swps))

back_V_fit = {} #Forgot to copy values sometimes, so using the Vsignal max value and R0 to estimate background voltage
K = {}

fig, ax = plt.subplots(1,figsize = (8,5))
for sample in swps:
    sweep = swps[sample]
    
    p0 =[sweep.argmin(),1e7,.5]
    fit_func, popt, sl = fit_lor(sweep, p0, window = 5)

    K[sample] = trmc.calc_K(f0 = popt[0], w = popt[1], R0 = popt[2], printparams = False)
    back_V_fit[sample] = -maxV[sample]*popt[2]
    
    ax.plot(sweep, marker = 'o', label = (sample + "\nK: " + str("{:0.4E}".format(K[sample]))))
    ax.plot(sweep.index.values[sl],fit_func[sl] , color = 'black')
    ax.set_xlabel('Frequency (GHz)')
    ax.set_ylabel('Normalized Reflectivity')
ax.legend()
ax.xaxis.get_offset_text().set_visible(False)
ax.set_xlim(8.42e9,8.652e9)
```

```python
direc = defaultpath + "Sample KCs\\Freq Sweep\\"

fps_freq =  {}

files = os.listdir(direc)
fnames = [f for f in files if os.path.isfile(os.path.join(direc,f))]
filepaths = [os.path.join(direc,fname) for fname in fnames]

amp, back_V = trmc.read_params(filepaths[0])

freq = [] 
regex = "^Freq=(\d+\.\d+)GHz"
for filepath in filepaths:
    fname = os.path.split(filepath)[1]
    m = re.search(regex, fname)
    freq.append(float(m.groups()[0])*1e9)

sortedfpaths = pd.Series(filepaths, index = freq)
sortedfpaths.sort_index(inplace = True)


data = trmc.read_trace(filepaths[0])
dfs_KCs = pd.DataFrame(index = data.index)

for fp in sortedfpaths:
    data = trmc.read_trace(fp, offsettime = 10e-9)
    dfs_KCs = pd.concat([dfs_KCs, data], axis = 1)
dfs_KCs.columns = sortedfpaths.index
```

```python
direc = defaultpath + "Sample A_R2\\Freq Sweep\\"

fps_freq =  {}

files = os.listdir(direc)
fnames = [f for f in files if os.path.isfile(os.path.join(direc,f))]
filepaths = [os.path.join(direc,fname) for fname in fnames]

amp, back_V = trmc.read_params(filepaths[0])

freq = [] 
regex = "^Freq=(\d+\.\d+)GHz"
for filepath in filepaths:
    fname = os.path.split(filepath)[1]
    m = re.search(regex, fname)
    freq.append(float(m.groups()[0])*1e9)

sortedfpaths = pd.Series(filepaths, index = freq)
sortedfpaths.sort_index(inplace = True)


data = trmc.read_trace(filepaths[0])
dfs_A = pd.DataFrame(index = data.index)

for fp in sortedfpaths:
    data = trmc.read_trace(fp, offsettime = 10e-9)
    dfs_A = pd.concat([dfs_A, data], axis = 1)
dfs_A.columns = sortedfpaths.index
```

```python
fig, axes = plt.subplots(2,1, figsize = (7,14), sharex = True)

sweep = swps['A_1000ms']

freq_i = np.linspace(sweep.index.min(), sweep.index.max(),50)


sweepV = sweep*(back_V/sweep.min())*(-1)
sweepV_i = pd.Series(np.interp(freq_i,sweepV.index,sweepV), index = freq_i)
swpmin = sweepV_i.argmin()

cuttimes =  dfs_A.index[::20]

dw = []
dfwhm = []
dR = []

timesel = 350e-9
timesel = min(cuttimes, key=lambda x:abs(x-timesel))

for cuttime in cuttimes:
    deltaV_i = pd.Series(np.interp(freq_i,dfs_A.loc[cuttime].index,dfs_A.loc[cuttime]), index = freq_i)
    sweepnew = sweepV_i+deltaV_i

    sweepV_i_norm = sweepV_i/sweepV_i.max()
    sweepnew_norm = sweepnew/sweepnew.max()

    p0 =[swpmin,1e7,0.5]
    fit_func_swp, popt_swp, sl = fit_lor(sweepV_i_norm, p0, window = 15)

    p0 =[swpmin,1e7,0.5]
    fit_func_new, popt_new, sl_new = fit_lor(sweepnew_norm, p0, window = 15)
    
    
    dw.append(popt_new[0]-popt_swp[0])
    dR.append(popt_new[2]-popt_swp[2])
    dfwhm.append(popt_new[1]-popt_swp[1])
    
    
    if cuttime == timesel:
        axes[0].plot(deltaV_i)

        axes[1].plot(sweepV_i_norm, label = 'cavity sweep', marker = 'o')
        axes[1].plot(freq_i[sl],fit_func_swp[sl], label = 'cavity sweep fit')
        axes[1].plot(sweepnew_norm, color = 'r', label = 'cavity sweep + deltaV' ,  marker = 'o')
        axes[1].plot(freq_i[sl_new],fit_func_new[sl_new], label = 'cavity sweep + deltaV fit')

        axes[1].axvline(swpmin, color = 'gray', linestyle = '--')
        axes[0].axvline(swpmin, color = 'gray', linestyle = '--')
        axes[0].axhline(0, color = 'gray', linestyle = '--')

        axes[0].set_ylabel('Delta V (V)')
        axes[1].set_ylabel('$P_r/P_rmax$')
        axes[1].set_xlabel('Frequency (Hz)')

        axes[1].set_xlim([8.525e9,8.545e9])
        axes[1].set_ylim([0.4,0.8])
        
        fig.suptitle('Delta V taken at ' + str(cuttime) + 's')

        plt.legend()

    
dw = pd.Series(dw, index = cuttimes)
dR = pd.Series(dR, index = cuttimes)
dfwhm = pd.Series(dfwhm, index = cuttimes)
```

```python
fig, axes = plt.subplots(3,1, figsize = (4,8), sharex = True)
# fig, ax = plt.subplots()

axes[0].plot(dw/max(dw))
axes[0].set_yscale('log')
axes[0].set_ylabel('$\Delta \omega$')

axes[1].plot(dfwhm/max(dfwhm))
axes[1].set_yscale('log')
axes[1].set_ylabel('$\Delta FWHM$')

axes[2].plot(dR/max(dR))
axes[2].set_yscale('log')
axes[2].set_ylabel('$\Delta R$')


axes[2].set_xlim(100e-9,1500e-9)
fmt = exp_formatter(-9)
axes[2].xaxis.set_major_formatter(FuncFormatter(fmt.format_func))
axes[2].set_xlabel('Time (ns)')
for ax in axes:
    ax.set_ylim([1e-3,2])
    
```
