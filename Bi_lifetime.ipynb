## Downselect data 

```python
fluencesel = slice(2,-1)
# fluencesel = -1
direc = 'U'

def fluencered(da):
    da_s = da.dropna('fluence','all').isel(fluence = fluencesel).dropna('time','all')
    return da_s

sigr_s = sigr.groupby('sample').apply(fluencered)
sigi_s = sigi.groupby('sample').apply(fluencered)
```

```python
g = sigr_s.plot(row = 'direction', col = 'sample',x = 'time', hue = 'fluence')
plt.yscale('log')
g.fig.legends[0].remove()
g.fig.autofmt_xdate()
# plt.xlim([400e-9,600e-9])
```

## Linear fitting of $log(\sigma)$

```python
sigi_ln = np.log(sigi_s)
sigr_ln = np.log(sigr_s)
# gb = sigi_ln.groupby('fluence','direction')

idxs = {'sample' : sigi_ln.indexes['sample'],'fluence' : sigi_ln.indexes['fluence'],'direction' : sigi_ln.indexes['direction']}

seldicts = list(load.dict_product(idxs))

seldicts_red = []
for i, seldict in enumerate(seldicts):
    t = sigi_ln.sel(seldict).dropna('time').to_series()
    if len(t) != 0:
        seldicts_red.append(seldict)
        
seldicts = seldicts_red
```

### Imaginary

```python
da_fits = sigi_ln.where(False).copy()
da_p = sigi_ln.isel(time = 0).drop('time').where(False).copy()

for seldict in seldicts:
    d = sigi_ln.sel(seldict).dropna('time').to_series()
    dsel = d.loc[800e-9:1200e-9]
    
    p = np.polyfit(dsel.index,dsel.values,1)
    
    da_p.loc[seldict] = p[0]
    f = np.poly1d(p)
    fit = pd.Series(f(dsel.index), index = dsel.index)
    tempseldict = seldict.copy()
    tempseldict['time'] = dsel.index
    da_fits.loc[tempseldict]  = fit
```

```python
fig,axes = plt.subplots(2,len(sigi_s.indexes['sample']), figsize = (20,10))

colsel = {}
for i, item in enumerate(sigi_s.indexes['sample']):
    colsel[item] = i
    
rowsel = {}
for i, item in enumerate(sigi_s.indexes['direction']):
    rowsel[item] = i

for seldict in seldicts:
    col = colsel[seldict['sample']]
    row = rowsel[seldict['direction']]
    data = sigi_ln.sel(seldict).to_series()
    axes[row][col].plot(data)
    
for ax in axes.flatten():
    ax.xaxis.set_major_formatter(FuncFormatter(plot.exp_formatter(-9).func))
    lns = ax.lines
    for i, ln in enumerate(lns):
        color = (0,i/len(lns),i/len(lns))
        ln.set_color(color)
        
for seldict in seldicts:
    col = colsel[seldict['sample']]
    row = rowsel[seldict['direction']]
    fit = da_fits.sel(seldict).to_series()
    axes[row][col].plot(fit, color = 'r')
```

```python
da_p.name = 'slope'
g = da_p.plot(x = 'fluence', col = 'direction', hue = 'sample')

for ax in g.axes.flatten():
    lns = ax.lines
    for ln in lns:
        plot.dropna_ln(ln)
    ax.set_ylim([-1e7,0])
```

## Look at time required for fractional decay

```python
da_dt = sigr_s.isel(time = 0).drop('time').where(False).copy()

frac = 10

for seldict in seldicts:
    d = sigr_s.sel(seldict).dropna('time').to_series()
    dmax = d.max()
    tmax = d.idxmax()
    dred = d.loc[tmax:]
    dred.plot()
    
    dfrac = dmax/frac
    tfrac = abs(dred-dfrac).idxmin()
    dt = tfrac-tmax
    da_dt.loc[seldict] = dt


plt.yscale('log')
```

```python
g = da_dt.plot(x = 'fluence', col = 'direction', hue = 'sample')

print('frac = ' + str(frac))
for ax in g.axes.flatten():
    lns = ax.lines
    for ln in lns:
        plot.dropna_ln(ln)
#     ax.set_ylim([-1e7,0])

plt.yscale('log')
plt.xscale('log')
```

## Max conductivity 

```python
def fluencenorm(da):
    fluence = da.coords['fluence'].values.item()
    return da/fluence

da = sigi_s.max('time').groupby('fluence').apply(fluencenorm)

g = da.plot(x = 'fluence', col = 'direction', hue = 'sample')


for ax in g.axes.flatten():
    lns = ax.lines
    for ln in lns:
        plot.dropna_ln(ln)
```
